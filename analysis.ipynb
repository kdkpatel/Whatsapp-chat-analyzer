```python
import pandas as pd
import re
```

```python
f = open('_chat.txt','r',encoding='utf-8')
```

```python
# reading data of the file in the string format
data = f.read()
```

```python
# print(data)
```

```python
pattern = r'\[(\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}:\d{2} [AP]M)\]\s(.*?)(?=\n\[\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}:\d{2} [AP]M\]|\Z)'

```

```python
matches = re.findall(pattern, data)
dates = [match[0] for match in matches]
messages = [match[1].strip() for match in matches]

```

```python
dates
```

```python
messages
```

```python
# Ensure both lists have the same length
assert len(dates) == len(messages), "Length mismatch between dates and messages."

'''This line checks if the lengths of the dates and messages lists are equal using the assert statement.
If the lengths are not equal, it raises an AssertionError with the specified message.'''
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %I:%M:%S %p')
df.rename(columns={'message_date': 'Date'}, inplace=True)
```

```python
def fetch_stats(selected_user,df):

    if selected_user == 'Overall':
        # 1. fetch number of messages
        num_messages = df.shape[0]

        # 2. fetch number of words
        words = []
        for message in df['Message']:
            words.extend(message.split())
        return num_messages, len(words)
    else:
        new_df = df[df['User'] == selected_user]
        num_messages = new_df.shape[0]
        words = []
        for message in new_df['Message']:
            words.extend(message.split())
        return num_messages, len(words)
```

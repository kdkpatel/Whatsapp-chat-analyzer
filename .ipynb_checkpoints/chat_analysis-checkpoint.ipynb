```python
import pandas as pd
import re
```

```python
f = open('_chat.txt','r',encoding='utf-8')
```

```python
# reading data of the file in the string format
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = r'\[(\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}:\d{2} [AP]M),]'
```

```python
matches = re.findall(pattern, data)
```

```python
dates = [match for match in matches]
```

```python
messages = re.split(pattern, data)[1:]
messages = [msg.strip() for msg in messages]
```

```python
messages
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df['message_date'] = pd.to_datetime(df['message_date'], format='%y/%m/%d, %I:%M:%S %p')
df.rename(columns={'message_date': 'date'}, inplace=True)
```
